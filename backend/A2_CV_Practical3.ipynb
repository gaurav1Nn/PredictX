#### Aim:Apply filters such as Gaussian,Median and Sobel on images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('premium_photo-1681400285322-048eeaa7f1c1.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)
median_blur = cv2.medianBlur(image, 5)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y).astype(np.uint8)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title('Gaussian Blur')
plt.imshow(gaussian_blur)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title('Median Blur')
plt.imshow(median_blur)

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))

plt.title('Sobel Combined')
plt.imshow(sobel_combined)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Sobel X')
plt.imshow(np.abs(sobel_x).astype(np.uint8))

```

```python
plt.figure(figsize=(10, 6))

plt.title('Sobel Y')
plt.imshow(np.abs(sobel_y).astype(np.uint8))
```
